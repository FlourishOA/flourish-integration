```python
import requests
import pandas as pd
import sys
import json

import sys
stdout = sys.stdout
reload(sys)
sys.setdefaultencoding('utf-8')
sys.stdout = stdout
```

```python
df  = pd.read_csv("data/doaj.csv")
```

```python
url = "http://flourishoa.org"
journal_ep = url + "/api/journals/"
price_ep = url + "/api/prices/"
```

```python
# authenticate to api
r = requests.post(url + '/api-token-auth/',
                  data={
                      'username': '',
                      'password': '',
                  })
token = "Token " + json.loads(r.text)['token']
print token
issns = []
```

```python
for index, row in df.iterrows():
    # make request to journals
        # 'Journal title', 'Journal EISSN (online version)'
        # 'Publisher',  'Subjects'
    issn = row['Journal EISSN (online version)']
    category = str(row['Subjects']).split(":")[0]
    
    journal_request_data = {
        u"issn": row['Journal EISSN (online version)'], 
        u"category": category,
        u"pub_name": unicode(row['Publisher']),
        u'journal_name': unicode(row['Journal title']),
        u'is_hybrid': False,
        u'url': row['Journal URL'],
    }
    issns.append(issn)
    #print(category)
    
    journal_request = requests.put(
        journal_ep + str(issn) + "/",
        headers={
            'Authorization': token,
            'Content-Type': 'application/json',
        },
        data=json.dumps(journal_request_data),
    )
    
    j_status = journal_request.status_code
    if j_status != 201 and j_status != 200:
        print str(issn) + ": " + str(j_status) + " :: " + str(row['Subjects'])
        
    
```

```python
# 5/22/17 at 7 PM PST
exchange_rates = {
    u'EUR': 1.12,
    u'PKR': 0.0095,
    u'COP': 0.00034, 
    u'USD': 1, 
    u'JPY': 0.0090, 
    u'CHF': 1.028, 
    u'IDR': 0.000075, 
    u'INR': 0.015, 
    u'ARS': 0.062, 
    u'MXN': 0.054, 
    u'CZK': 0.042, 
    u'BRL': 0.301, 
    u'GBP': 1.300, 
    u'CNY': 0.145, 
    u'ZAR': 0.076, 
}
```

```python
for index, row in df.iterrows():
    if index < 21:
        continue
    issn = row['Journal EISSN (online version)']
    apc_lab = row['Journal article processing charges (APCs)']
    apc = -1
    license_info = 'Unknown'
    if apc_lab == 'No Information':
        continue
    elif apc_lab == 'No':
        apc = 0
    else:
        # get the information from 
        issn = row['Journal EISSN (online version)']
        
        # request the data from DOAJ
        
        doaj_api_ep = 'https://doaj.org/api/v1/search/journals/issn%3A'
        
        doaj_request = requests.get(
            doaj_api_ep + str(issn),
            headers = {'Accept': 'application/json',}
        )
    
        doaj_status = doaj_request.status_code
        if doaj_status == 200:
            res = doaj_request.json()
            if res['results']:
                license_info = res['results'][0]['bibjson']['license'][0]['type']
                if license_info[0:3] == 'Pub':
                    license_info = 'Standard Copyright'
                elif license_info[0:3] == 'Any':
                    license_info = 'CC0'
                apc_info = res['results'][0]['bibjson']['apc']
                curr = apc_info['currency'] 
                if curr not in exchange_rates:
                    print curr + ": wasn't found"
                    continue
                else:
                    apc = round(apc_info['average_price'] \
                                * exchange_rates[curr], 2)
    
    date = row['Added on Date'].split("T")[0]
        
    price_request_data = {
        "issn": row['Journal EISSN (online version)'], 
        "date_stamp": date,
        'license': license_info,
        'url': row['Journal URL'],
        'price': apc,
    }

    price_request = requests.put(
        price_ep + str(issn) + "/",
        headers={
            'Authorization': token,
            'Content-Type': 'application/json',
        },
        data=json.dumps(price_request_data),
    )
    
    p_status = price_request.status_code
    print str(issn) + ": " + str(p_status)
```
